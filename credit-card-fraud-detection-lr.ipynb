 # Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and Fradulent transactions.

Number of Fradulent transcations is 492

```python
legit_sample = legit.sample(n=492,random_state=2)
legit_sample.shape
```

```python
# now concatenating two DataFrames 
new_dataset = pd.concat([legit_sample,fraud],axis=0) # concat rows wise
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

here, we can see that the nature of the new_dataset and original dataset are almost same (familiar) and got good samples.

# Spliting the data into Features and target

```python
x = new_dataset.drop(['Class'],axis=1)
y = new_dataset['Class']
```

```python
print(x.shape)
print(y.shape)
```

# Split the data into training and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, stratify=y, random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

# Model training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit(x_train,y_train)
```

Model Evaluation

```python
# To check the accuracy score on training data
y_train_pred = model.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score

print('model accuracy score on training data : {:.4f}'.format(accuracy_score(y_train_pred,y_train)))
```

```python
# Check the accuracy score on test data
y_pred = model.predict(x_test)

print('model accuracy score on test data : {:.4f}'.format(accuracy_score(y_pred,y_test)))
```

* The accuracy score on training data (0.9149) and accuracy score on test data (0.9137) are same. So, there is no underfitting and overfitting problem in this is model.
* The prediction of this model is very accurate on both training and test data. 
* so, this is the best model for predictions of 'credit card fraud detection'.
